```python
import numpy as np
```

```python
std_marks = np.genfromtxt('students.txt',delimiter=',',skip_header=1)
print(std_marks)
```

```python
# Displaying marks of student with id 1,2,3
print(f'Marks of student with id {std_marks[0,0].astype(int)} is {std_marks[0,1:]}')
print(f'Marks of student with id {std_marks[1,0].astype(int)} is {std_marks[1,1:]}')
print(f'Marks of student with id {std_marks[2,0].astype(int)} is {std_marks[0,1:]}')
```

```python
#Print the total marks obtained by each student in all subjects (out of 225)
total_marks = np.sum(std_marks,axis=1)
print(f'Total marks of student with id [{std_marks[0,0].astype(int)},{std_marks[1,0].astype(int)},{std_marks[2,0].astype(int)}] is {total_marks[:]} ')
```

```python
#Print the highest marks scored by each student along with subject name
print(f'Highest marks of student with id [{std_marks[0,0].astype(int)},{std_marks[1,0].astype(int)},{std_marks[2,0].astype(int)}] is {np.max(std_marks,axis=1)}')
```

```python
#Print the highest marks scored by student in each subject
print(f'Highest marks on each subject is  {np.max(std_marks,axis=0)}')
```

```python
std_marks1 = std_marks[:,1:]
print(std_marks1)
```

```python
#Average marks of students in each subject is 
print(np.average(std_marks1,axis=0).round(1))
```
